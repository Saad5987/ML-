# My Lane as an ML Task

## Selected Lane
Content Optimization

## ML Task Type
Classification

I selected the Content Optimization lane because the goal is to identify webpages that should be prioritized for optimization. This is a classification problem because each webpage can be classified as either requiring optimization or not requiring immediate optimization. The model will help content teams make better decisions about which pages to improve first.

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("content_refresh_anonymized.csv")

# Display dataset shape
print("Dataset Shape:", df.shape)
```

# Target or Proxy

The dataset does not contain a direct label indicating whether a webpage needs optimization. Therefore, a proxy target is created.

A webpage is considered a priority if:

- The trend direction is **down**
- The average search position is greater than **10**

This proxy helps represent pages that may benefit from SEO or content improvements.

```python
# Create a proxy target column

df["priority"] = (
    (df["trend_direction"] == "down") &
    (df["avg_position"] > 10)
).astype(int)

df[["trend_direction", "avg_position", "priority"]].head(10)
```

# Success Metric

The primary success metric will be the **F1 Score** because it balances precision and recall when identifying pages that should be prioritized for optimization.

Accuracy will also be reported to measure overall model performance.

In practice, business metrics such as improvements in CTR or organic traffic after content updates can also be monitored.

# Unit of Analysis

The unit of analysis is **one webpage**.

# Why ML Beats a Fixed Rule

A simple rule such as "optimize every page with low CTR" ignores many other important factors.

Webpage performance depends on multiple variables including search position, impressions, content age, trend direction, engagement rate, and content type.

Machine learning can learn patterns across these variables and provide more accurate recommendations than a single fixed rule. The model supports decision-making rather than replacing human judgment.
